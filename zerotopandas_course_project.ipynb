## Data Preparation and Cleaning

I had retrieved data only of 1 month from the data frame and aggregated sum of all days in that month. I had removed unnecessary columns and replaced NAN by 0.

```python
import pandas as pd
import numpy as np
from urllib.request import urlretrieve
import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns

urlretrieve('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv', 'owid-covid-data.csv')
covid_df = pd.read_csv('owid-covid-data.csv')
covid_df2 = covid_df.copy()
covid_df3 = covid_df.copy()
```

```python
covid_df
```

```python
covid_df = covid_df[covid_df.date >= '2020-01-01']
covid_df = covid_df[covid_df.date <= '2020-01-31']
```

```python
covid_df
```

```python
covid_df = covid_df[['continent', 'location', 'date', 'total_cases', 'new_cases',
       'new_cases_smoothed', 'total_deaths', 'new_deaths',
       'new_deaths_smoothed', 'new_tests', 'total_tests',
       'new_tests_smoothed',
       'population', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand']]
```

```python
covid_df.continent.fillna('International')
covid_df.fillna(0)
```

## Exploratory Analysis and Visualization

I had plotted graph to show how contagious Corona virus is among different age groups and who are more prone to this virus-Cardiovasc Death Rate, Diabetes Prevalence. I had shown total cases, total deaths, total tests etc via pie chart, gdp per capita, handwashing facilities, hospital beds per thousand through several types of graphs.

```python
ttlmonth_df  = covid_df.groupby('location').sum()
ttlmonth_df['location'] = ttlmonth_df.index
sns.set_style('whitegrid')
plt.figure(figsize=(20,5))
plt.plot(ttlmonth_df.location, ttlmonth_df.median_age, 's-r')
plt.plot(ttlmonth_df.location, ttlmonth_df.aged_65_older, 'o--g')
plt.plot(ttlmonth_df.location, ttlmonth_df.aged_70_older, '*:b')
plt.ylabel('No. of people')
plt.legend(['median_age', 'aged_65_older', 'aged_70_older'])
plt.xticks(rotation=80)
plt.show()
```

```python
plt.figure(figsize=(20,5));
plt.plot(ttlmonth_df.location, ttlmonth_df.cardiovasc_death_rate);
plt.plot(ttlmonth_df.location, ttlmonth_df.diabetes_prevalence);
plt.xticks(rotation=80);
```

```python
plt.figure(figsize=(20,5))
sns.barplot(ttlmonth_df.location, ttlmonth_df.gdp_per_capita)
plt.xticks(rotation=80)
plt.show()
plt.figure(figsize=(20,5))
sns.barplot(ttlmonth_df.location, ttlmonth_df.hospital_beds_per_thousand)
plt.xticks(rotation=80)
plt.show()
```

```python
smk_counts = ttlmonth_df[['female_smokers','male_smokers']].sum()
case_counts = ttlmonth_df[['total_cases', 'new_cases', 'new_cases_smoothed']].sum()
death_counts = ttlmonth_df[['total_deaths', 'new_deaths', 'new_deaths_smoothed']].sum()
test_counts = ttlmonth_df[['new_tests', 'total_tests', 'new_tests_smoothed']].sum()


plt.pie(smk_counts, labels=['female_smokers','male_smokers'], radius=2, autopct="%.2f%%", colors=['#00ff00', '#ff00ff'], explode=(.5, 0), shadow=True);
plt.show()

plt.pie(case_counts, radius=3, autopct="%.2f%%",  pctdistance=2.6/3, colors=['green', '#2040ff', '#ff5050']);
plt.pie(death_counts, radius=2, autopct="%.2f%%",  pctdistance=1.6/2, colors=['#6080ff', '#ff00ff', '#00ff00']);
plt.pie(test_counts, autopct="%.2f%%", colors=['#00ffff', 'yellow', 'red']);
plt.legend(['total_cases', 'new_cases', 'new_cases_smoothed', 'total_deaths', 'new_deaths', 'new_deaths_smoothed', 'new_tests', 'total_tests', 'new_tests_smoothed'], loc=(1,1));
```

```python
plt.figure(figsize=(24,8))
results = np.array(ttlmonth_df.handwashing_facilities).reshape(10,7)
strings = np.array(ttlmonth_df.location).reshape(10,7)
labels = (np.asarray(["{0} {1:.2f}".format(string, value)
                    for string, value in zip(strings.flatten(),
                                        results.flatten())]).reshape(10,7))
sns.heatmap(results, annot=labels, fmt='s', cmap=sns.color_palette("husl", 9), xticklabels='', yticklabels='');
plt.title('handwashing_facilities');
```
